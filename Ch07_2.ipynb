```python
from google.colab import drive
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as plt

drive.mount('/content/drive')
```

```python
img=cv2.imread('/content/drive/MyDrive/IP/Images/lena256.jpg')

print(img.shape)
gImg1=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#plt.imshow(gImg1,cmap='gray')
cv2_imshow(gImg1)
```

```python
M = cv2.getRotationMatrix2D((128,128),45,1)
outImg = cv2.warpAffine(gImg1,M,(256,256))
cv2_imshow(outImg)
```

```python
M = np.float32([[1,0,100],[0,1,50]])
outImg=cv2.warpAffine(gImg1,M,(256,256))
cv2_imshow(outImg)
```

```python
M = cv2.getRotationMatrix2D((128,128),-45,1)
outImg = cv2.warpAffine(gImg1,M,(256,256))
cv2_imshow(outImg)
```

```python
M1 = np.float32([[1,0,-128],[0,1,-128]])
outImg1=cv2.warpAffine(gImg1,M1,(256,256))
cv2_imshow(outImg1)
```

```python
print(M1)
```

```python
M2 = cv2.getRotationMatrix2D((0,0),-45,1)
outImg2 = cv2.warpAffine(outImg1, M2, (256,256))
cv2_imshow(outImg2)
```

```python
print(M2)
```

```python
M3 = np.float32([[1,0,128],[0,1,128]])
outImg3=cv2.warpAffine(outImg2,M3,(256,256))
cv2_imshow(outImg3)
```

```python
print(M3)
```

```python
N1 = np.float32([[M1[0,0],M1[0,1],M1[0,2]],[M1[1,0],M1[1,1],M1[1,2]],[0,0,1]])
N2 = np.float32([[M2[0,0],M2[0,1],M2[0,2]],[M2[1,0],M2[1,1],M2[1,2]],[0,0,1]])
N3 = np.float32([[M3[0,0],M3[0,1],M3[0,2]],[M3[1,0],M3[1,1],M3[1,2]],[0,0,1]])
print(N1)
```

```python
N4 = np.dot(N3,N2)
N5 = np.dot(N4,N1)
N6 = np.float32([[N5[0,0],N5[0,1],N5[0,2]],[N5[1,0],N5[1,1],N5[1,2]]])
print(N6)
```

```python
outImg4=cv2.warpAffine(gImg1,N6,(256,256))
cv2_imshow(outImg4)
```
