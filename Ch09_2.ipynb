```python
from google.colab import drive
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as plt

drive.mount('/content/drive')
```

```python
img=cv2.imread('/content/drive/MyDrive/IP/Images/abcdef.bmp')

gImg1=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(img.shape)
plt.imshow(gImg1,cmap='gray')
cv2_imshow(img)
```

```python
numLabels, labels = cv2.connectedComponents(gImg1)
print(numLabels)
```

```python
print(labels.shape)
print(labels)
```

```python
label_H=np.uint8(labels*179/numLabels)
print(label_H.shape)
cv2_imshow(label_H)
```

```python
bank=255*np.ones_like(label_H)
labelImg=cv2.merge([label_H,bank,bank])
labelImg=cv2.cvtColor(labelImg,cv2.COLOR_HSV2RGB)
labelImg[label_H==0]=0
cv2_imshow(labelImg)
```

```python
img=cv2.imread('/content/drive/MyDrive/IP/Images/milkdrop.bmp')

gImg1=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(img.shape)
plt.imshow(gImg1,cmap='gray')
```

```python
ret, thresh = cv2.threshold(gImg1, 127, 255, 0)
cv2_imshow(thresh)
```

```python
contour, hierarchy=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
img1=cv2.drawContours(img,contour,-1,(0,0,255),1)
cv2_imshow(img1)
```
