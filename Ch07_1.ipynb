```python
from google.colab import drive
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as plt

drive.mount('/content/drive')
```

```python

img=cv2.imread('/content/drive/MyDrive/IP/Images/shrinking.bmp')

print(img.shape)
gImg1=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#plt.imshow(gImg1,cmap='gray')
cv2_imshow(gImg1)
```

```python
outImg=np.zeros(gImg1.shape, dtype=np.uint8)
#cv2_imshow(outImg)
```

```python
scaling=3
for y in range(gImg1.shape[0]):
    for x in range(gImg1.shape[1]):
        h=int(y/scaling)
        w=int(x/scaling)
        if h < gImg1.shape[0] and x < gImg1.shape[1]:
            outImg[y,x]=gImg1[h,w]

cv2_imshow(outImg)
```

```python
outImg=cv2.resize(gImg1, None, fx=0.3, fy=0.3, interpolation=cv2.INTER_NEAREST)
cv2_imshow(outImg)

#outImg=cv2.resize(gImg1, None, fx=0.3, fy=0.3, interpolation=cv2.INTER_LINEAR)
#outImg=cv2.resize(gImg1, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_CUBIC)
```

```python
blur=cv2.GaussianBlur(gImg1,(5,5),0)
outImg=cv2.resize(blur, None, fx=0.3, fy=0.3, interpolation=cv2.INTER_NEAREST)
cv2_imshow(outImg)
```

```python

img=cv2.imread('/content/drive/MyDrive/IP/Images/lena256.jpg')

print(img.shape)
gImg1=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#plt.imshow(gImg1,cmap='gray')
cv2_imshow(gImg1)
```

```python
height, width= gImg1.shape
print(height,width)
```

```python
M = cv2.getRotationMatrix2D((width/2.0, height/2.0),45,1)
print(M)
```

```python
outImg=cv2.warpAffine(gImg1,M,(width,height))
cv2_imshow(outImg)
```
