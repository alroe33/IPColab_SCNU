```python

img=cv2.imread('/content/drive/MyDrive/IP/Images/lena256.jpg')

print(img.shape)
gImg1=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gImg1,cmap='gray')
#cv2_imshow(gImg1)
```

로버츠 검추기

```python
gx = np.array([[-1,0],[0,1]], dtype=int)
gy = np.array([[0,-1],[1,0]], dtype=int)
```

```python
gx = np.array([[-1,0,0],[-2,0,2],[-1,0,1]], dtype=int)
gy = np.array([[1,2,1],[0,0,0],[-1,-2,-1]], dtype=int)
```

소벨 검출기

```python
gx = np.array([[-1,0,0],[-2,0,2],[-1,0,1]], dtype=int)
gy = np.array([[1,2,1],[0,0,0],[-1,-2,-1]], dtype=int)
```

프리윗 검출기

```python
gx = np.array([[-1,0,0],[-1,0,1],[-1,0,1]], dtype=int)
gy = np.array([[1,1,1],[0,0,0],[-1,-1,-1]], dtype=int)
```

엣지 검출

```python
x=cv2.filter2D(gImg1,-1,gx)
y=cv2.filter2D(gImg1,-1,gy)
```

```python
absx = cv2.convertScaleAbs(x)
absy = cv2.convertScaleAbs(y)
```

```python
outImg = cv2.addWeighted(absx,1,absy,1,0)

plt.imshow(absx,cmap='gray')
#cv2_imshow(outImg)
```

```python
plt.imshow(absy,cmap='gray')
#cv2_imshow(absx
```

```python
gBin = np.zeros(gImg1.shape, dtype=np.uint8)
```

```python
cv2.threshold = 40

for i in range(gImg1.shape[0]):
    for j in range(gImg1.shape[1]):
        if outImg[i,j] > cv2.threshold:
            gBin[i,j] = 255


plt.imshow(gBin,cmap='gray')
#cv2_imshow(gBin
```
