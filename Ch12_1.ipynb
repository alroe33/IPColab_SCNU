```python
from google.colab import drive
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as plt

drive.mount('/content/drive')
```

```python
img=cv2.imread('/content/drive/MyDrive/IP/Images/home3.jpg')

gImg1=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
print(img.shape)
plt.imshow(gImg1,cmap='gray')
#cv2_imshow(gImg1)
```

```python
img2=img.copy()
```

```python
edges=cv2.Canny(gImg1,100,200)
plt.imshow(edges,cmap='gray')
```

```python
lines=cv2.HoughLinesP(edges,1,np.pi/80,220)
```

```python


for line in lines:
  rho, theta =line[0]
  a=np.cos(theta)
  b=np.sin(theta)
  x0=a*rho
  y0=b*rho

  x1,y1=(int(x0+1000*(-b)),int(y0+1000*(a)))
  x2,y2=(int(x0-1000*(-b)),int(y0-1000*(a)))

  cv2.line(img2.copy(),(x1,y1),(x2,y2),(0,0,255),1)

plt.imshow(img2)
```
